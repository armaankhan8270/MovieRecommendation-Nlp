```python
df2=pd.read_csv('/kaggle/input/tmdb-movie-metadata/tmdb_5000_movies.csv')
df=pd.read_csv('/kaggle/input/tmdb-movie-metadata/tmdb_5000_credits.csv')
```

```python
df2=df2[['genres','keywords', 'overview', 'release_date', 'title']
         ]
```

```python
movies= df.merge(df2, on='title')
```

```python

import ast
def Convert(text):
    L=[]
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
```

```python
movies
```

```python
movies['genres']=movies['genres'].apply(Convert)
```

```python
movies['keywords']=movies['keywords'].apply(Convert)
```

```python

import ast
def Convert2(text):
    L=[]
    c=0
    for i in ast.literal_eval(text):
        c+=1
        if c!=4:
           L.append(i['name'])
        else:
            break
    return L
```

```python
movies['cast']=movies['cast'].apply(Convert2)
```

```python
movies['cast'][0]
```

```python
import ast

def director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
movies['crew']=movies['crew'].apply(director)
```

```python
def Li(text):
    if isinstance(text, str):
       return text.split()
    else:
       return text
def listToString(s):
    # Use str.join() to concatenate elements from the list into a string
    str1 = ''.join(str(ele) for ele in s if ele is not None)
    return str1

    
    
```

```python
movies['overview']=movies['overview'].apply(Li)
```

```python
movies['overview']=movies['overview'].apply(str)
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(str(i).replace(" ",""))
    return L1
```

```python
movies[['genres','cast','crew','keywords']]=movies[['genres','cast','crew','keywords']].apply(collapse)
```

```python
movies.sample(3)
```

```python
movies['tags']=movies['cast']+movies['crew']+movies['genres']+movies['keywords']+movies['overview']
movies['tags'][0]
```

```python
final_df=movies[['movie_id','title','tags']]
```

```python
def reco(text):
    text=text.replace("[", "").replace("]", "").replace("'", "").replace(',',' ')
   
    
    return ''.join(text)
```

```python
final_df['tags']=final_df['tags'].apply(reco)
```

```python
final_df['tags'][0]
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)    
```

```python
stem("loved loving")
```

```python
final_df['tags']=final_df['tags'].apply(stem)
```

```python
final_df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=3000,stop_words='english')
```

```python
vectors=cv.fit_transform(final_df['tags']).toarray()
```

```python
vectors
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
def recomed(movie):
    index = final_df[(final_df['title'] == movie) | (final_df['title'] == movie.title())].index[0]


    distance=similarity[index]
    movie_index=sorted(list(enumerate(similarity[index])),reverse=True,key=lambda x:x[1])[1:6]
    
    for m in movie_index:
        print(final_df.iloc[m[0]].title)

    return movie_index
```
